```

import os
files = []
os.chdir(os.path.join("data", "test"))
for name in os.listdir(os.getcwd()):
    if name.endswith(".jpg"):
        files.append("data/test/" + name)
os.chdir("..")
with open("test.txt", "w") as f:
    for img in files:
        f.write(img)
        f.write("\n")
    f.close()
os.chdir("..")
```

```
import os
files = []
os.chdir(os.path.join("data", "obj"))
for name in os.listdir(os.getcwd()):
    if name.endswith(".jpg"):
        files.append("data/obj/" + name)
os.chdir("..")
with open("train.txt", "w") as f:
    for img in files:
        f.write(img)
        f.write("\n")
    f.close()
os.chdir("..")
```
